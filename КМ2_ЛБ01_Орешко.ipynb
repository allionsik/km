*Подход 1* **(точное решение):** время касания телом земли находим как точное решение уравнения 
$
s_{y}(t) = 0;
$

*Подход 2* **(приближенное решение):**

- Постройте последовательность $s_{y}$ на временном отрезке $[0, t_{\text{stop}}]$, чтобы момент касания тела земли $t^{*}$ попадал в выбранный отрезок, т.е. $t^{*} \in [0, t_{\text{stop}}]$. Отметим, что $t_{\text{stop}} > T$, где значение $T$ задается в Задании 1.1.  
- Найдите в построенной последовательности $s_{y}$ два соседних элемента последовательности, значения которых меняют знак. Т.е. найдите значение индекса $k$, для которого $
  s_{y}[k] \cdot s_{y}[k+1] < 0.
  $
  Это будет означать, что:  
  $
  0 \in (s_{y}[k], s_{y}[k+1]).
  $  
- Постройте прямую линию $s_{\text{lin}}(t)$, проходящую через две точки $(t[k], s_{y}[k])$ и $(t[k+1], s_{y}[k+1])$, и найдите $t^{*} \in (t[k], t[k+1])$ из условия $s_{\text{lin}}(t^{*}) = 0$. Найденное значение $t^*$, будет соответствовать точке $(t^*, 0)$ на прямой $s_{lin}(t)$. $\textcolor{red}{\text{Приведите}}$ в документе формулы для прямой $s_{lin}(t)$ и для вычисления $t^*$.
$\textcolor{red}{\text{Сравните}}$ точное решение *(Подход 1)* и приближенное решение *(Подход 2)*, вычислив относительную ошибку.

**Построение касательной.** Цвет построения: голубой

Касательная задается уравнением вида: $y = k(x-x_{\text{start}})+y_{\text{start}}$

Угловой коэффициент находится по формуле: $ k=\frac{y'(t)}{x'(t)} $ , при $t=0$

```python
s_x = v0_x * t  
s_y = h_start + v0_y * t + g * t**2 / 2 
t = sympy.symbols('t')
difx = sympy.diff(s_x, t)
dify = sympy.diff(s_y, t)
k = dify.subs(t, 0) / difx.subs(t, 0)
#направляющий вектор по двум точкам
y_kas1 = np.array([s_x_array[0], s_y_array[0]])
y_kas2 = np.array([s_x_array[0] + 1, k + s_y_array[0]])  
y_kas = [y_kas1 + (y_kas2 - y_kas1) * t for t in [-1, 2]] 
y_kas = np.array(y_kas)
plt.plot(s_x_array,s_y_array,'lightgreen',lw=3)
plt.plot(p[:,0],p[:,1],'-',color='pink',lw=3) 
plt.plot(y_kas[:,0],y_kas[:,1],'-',color='lightblue',lw=3) 
plt.xlabel('x (m)'); plt.ylabel('y (m)')
plt.axis([-5, s_end+5, 0, h_end+11]);
```

**Построить график нормали.** Цвет построения:оранжевый

Свойство: Произведение угловых коэффициентов двух перпендикулярных прямых равно -1. $k_{\text{касательной}}*k_{\text{нормали}}=-1$. => $k_{\text{нормали}}=\frac{-1}{k_{\text{касательной}}}$

Уравнение нормали: $y = k(x-x_{\text{start}})+y_{\text{start}}$

```python
y_normal1 = np.array([s_x_array[0], s_y_array[0]])
y_normal2 = np.array([s_x_array[0] + 1, -1 / k + s_y_array[0]])
y_norm = [y_normal1 + (y_normal2 - y_normal1) * t for t in [-5, 10]]
y_norm = np.array(y_norm)
s_x = v0_x * t  
s_y = h_start + v0_y * t + g * t**2 / 2 
t = sympy.symbols('t')
difx = sympy.diff(s_x, t)
dify = sympy.diff(s_y, t)
k = dify.subs(t, 0) / difx.subs(t, 0)
#направляющий вектор по двум точкам
y_kas1 = np.array([s_x_array[0], s_y_array[0]])
y_kas2 = np.array([s_x_array[0] + 1, k + s_y_array[0]])  
y_kas = [y_kas1 + (y_kas2 - y_kas1) * t for t in [-1, 2]] 
y_kas = np.array(y_kas)
plt.plot(s_x_array,s_y_array,'lightgreen',lw=1)
plt.plot(p[:,0],p[:,1],'-',color='pink',lw=1) 
plt.plot(y_kas[:,0],y_kas[:,1],'-',color='lightblue',lw=1) 
plt.plot(y_norm[:,0],y_norm[:,1],'-',color='orange',lw=1)
plt.xlabel('x (m)'); plt.ylabel('y (m)')
plt.axis([-5, s_end+5, 0, h_end+11]);
```
